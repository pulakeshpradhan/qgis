# PyQGIS: Mastering GIS Automation

Welcome to the ultimate guide for automating QGIS with Python and Jupyter. This book will take you from a beginner to an advanced PyQGIS developer.

## What is PyQGIS?

PyQGIS is the Python API for QGIS. It allows you to:
- **Automate** repetitive GIS tasks.
- **Extend** the functionality of QGIS by creating plugins.
- **Integrate** QGIS with other Python libraries (Pandas, Numpy, Scikit-learn).
- **Run** analysis without even opening the QGIS interface (Standalone scripts).

## How to use this book

This book is organized into sections:

- **Fundamentals**: Getting your environment ready.
- **Vector Operations**: Data manipulation, attributes, and geometry.
- **Raster Operations**: Band math and spatial analysis.
- **Workflow Automation**: Using the Processing Toolbox and batch scripts.
